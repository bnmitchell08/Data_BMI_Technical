```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ds1 = pd.read_csv('dataset1.csv')
ds2 = pd.read_csv('dataset2.csv')
#ds2.head(20)
merged = pd.merge(ds1, ds2, on='record_id')

data_types = merged.dtypes
print(data_types)
output_file_path = 'merged.csv'
merged.to_csv(output_file_path, index=True)
```

```python
merged.rename(columns={'Sex':'Gender'}, inplace=True)
merged.to_csv('merged.csv', index=False)
merged.head(15)
```

```python
#Convert Height (cm) to Height (m)
#Add column for height in meters  cm/100= meter
merged['Height (m)'] = merged['Height (cm)'] / 100 

#Add column for BMI = Weight(kg)/height(m)^2
merged['BMI'] = merged['Weight (kg)'] / (merged['Height (m)'] ** 2)
merged.head(5)

num_records = merged.shape[0]
print("Number of records in the dataset:", num_records)
merged.head(15)
```
