```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

merged = pd.read_csv('merged_BMI.csv')

plt.figure(figsize=(10,6))
counts, bins, _ = plt.hist(merged['BMI'], bins = 25, color = 'lightblue', edgecolor='grey')
plt.xlabel('BMI')
plt.ylabel('Distribution of BMI')
plt.title("Distribution of BMI from Merged Patient Dataset")

#Annotation
for i in range(len(counts)):
    plt.text(bins[i]+.6, counts[i] + 1, str(int(counts[i])), ha='center', va='bottom')
plt.savefig('BMI_Distribution_Histo.png',dpi=300, bbox_inches='tight')
```

```python
bmi_underweight = merged[(merged['BMI']<=18.5)]
bmi_healthyweight = merged[(merged['BMI']>=18.6) & (merged['BMI']<=24.9)]
bmi_overweight = merged[(merged['BMI']>=25.0) & (merged['BMI']<=29.9)]
bmi_obesity = merged[(merged['BMI']>=30.0)]

bmi_underweight_grouped = bmi_underweight.groupby('Race').size().reset_index(name='Underweight')
bmi_healthyweight_grouped = bmi_healthyweight.groupby('Race').size().reset_index(name='Healthy Weight')
bmi_overweight_grouped = bmi_overweight.groupby('Race').size().reset_index(name='Overweight')
bmi_obesity_grouped = bmi_obesity.groupby('Race').size().reset_index(name='Obesity')

merged_bmi_groups = bmi_underweight_grouped.merge(bmi_healthyweight_grouped, on='Race', how='outer') \
                                           .merge(bmi_overweight_grouped, on='Race', how='outer') \
                                           .merge(bmi_obesity_grouped, on='Race', how='outer')

print(bmi_underweight_grouped)
print(merged_bmi_groups)

fig, ax = plt.subplots(figsize=(10, 6))
colors=['seagreen','turquoise', '#a3c585', '#c7ddb5']
# Plotting the stacked bars
ax.bar(merged_bmi_groups['Race'], merged_bmi_groups['Obesity'], bottom=merged_bmi_groups['Underweight'] + merged_bmi_groups['Healthy Weight'] + merged_bmi_groups['Overweight'], label='Obesity BMI 30.0+', color=colors[3])
ax.bar(merged_bmi_groups['Race'], merged_bmi_groups['Overweight'], bottom=merged_bmi_groups['Underweight'] + merged_bmi_groups['Healthy Weight'], label='Overweight BMI: 25.0 – 29.9', color=colors[2])
ax.bar(merged_bmi_groups['Race'], merged_bmi_groups['Healthy Weight'], bottom=merged_bmi_groups['Underweight'], label='Healthy Weight BMI: 18.5 – 24.9	', color=colors[1])
ax.bar(merged_bmi_groups['Race'], merged_bmi_groups['Underweight'], label='Underweight BMI: Below 18.5' ,color=colors[0])

plt.title('BMI Count by Race', fontsize = 16)
plt.xlabel('Race')
plt.ylabel('Number of Patients')
plt.legend(title='BMI Category', loc='upper left')
plt.xticks(rotation=45)  

#plt.show()
plt.savefig('BMI_count_by_race_bar.png',dpi=300, bbox_inches='tight')
```
